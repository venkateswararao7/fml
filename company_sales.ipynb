```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import linear_model
print("set up completed")
```

```python
file_path='/content/Company_data.csv'
file=pd.read_csv(file_path,parse_dates=True)
```

```python
plt.figure(figsize=(15,6))
plt.title("company sales")
sns.lineplot(data=file)
```

```python
X=file.drop("Sales",axis=1).values
y=file["Sales"].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=50)
```

```python
y_prediction=linear_model.LinearRegression()
y_prediction.fit(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_train_prediction=y_prediction.predict(x_train) 
print("mean_absolute_error = ",mean_absolute_error(y_train,y_train_prediction))
print("r2_score = ",r2_score(y_train,y_train_prediction))
print("mean square error = ",mean_squared_error(y_train,y_train_prediction))
print("error that are  found in train data(seen data )")
```

```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
y_test_prediction=y_prediction.predict(x_test) 
print("mean_absolute_error = ",mean_absolute_error(y_test,y_test_prediction))
print("r2_score = ",r2_score(y_test,y_test_prediction))
print("mean square error = ",mean_squared_error(y_test,y_test_prediction))
print("unseen data prediction")
```
